```python
data.drop("Cancelled Rides by Customer", axis=1, inplace=True)
# data.drop("Reason for cancelling by Customer", axis=1, inplace=True)
data.drop("Cancelled Rides by Driver", axis=1, inplace=True)
# data.drop("Driver Cancellation Reason", axis=1, inplace=True)
data.drop("Incomplete Rides", axis=1, inplace=True)
data.drop("Incomplete Rides Reason", axis=1, inplace=True)
data.drop("Booking ID", axis=1, inplace=True)
data.shape
```

```python
data["Avg VTAT"].fillna(data["Avg VTAT"].mean(), inplace=True)
avg_vtat= data.groupby("Vehicle Type")["Avg VTAT"].mean()
print(avg_vtat)

avg_vtat.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Average VTAT by Vehicles")
plt.xlabel("Vehicle Type")
plt.ylabel("Average VTAT")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
data["Avg CTAT"].fillna(data["Avg CTAT"].mean(), inplace=True)
avg_ctat= data.groupby("Vehicle Type")["Avg CTAT"].mean()
print(avg_ctat)

avg_ctat.plot(kind="bar", color="green", edgecolor="black")
plt.title("Average CTAT by Vehicles")
plt.xlabel("Vehicle Type")
plt.ylabel("Average CTAT")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="Dark2", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
common_count=data["Pickup Location"].value_counts().head(100).sum()
uncommon_count=data["Pickup Location"].value_counts().sum()-common_count
chart=[common_count, uncommon_count]
labels = ["Top 100 Common Locations", "Other Uncommon Locations"]
plt.pie(chart, labels=labels, autopct="%1.1f%%", shadow=True, wedgeprops={'edgecolor': 'black'})
plt.title("Common vs Uncommon Pickup Locations")
plt.show()
```

Top 100 most common locations contitute about "58%" of total ride bookings.


```python
bk_status= data["Booking Status"].value_counts()
bk_status_df = data["Booking Status"].value_counts().reset_index()
bk_status_df.columns=["Booking status", "Count"]
print(bk_status_df)
```

```python
plt.xticks(rotation=30, ha="right")
plt.xlabel("Booking Status")
plt.ylabel("Count Of Status")
plt.bar(bk_status.index, bk_status.values, color="#FF7518")
```

```python
plt.style.use("fivethirtyeight")
status_count=bk_status
status=["Completed", "Cancelled by Driver", "No Driver Found", "Cancelled by Customer", "Incomplete"]
explode=[0, 0, 0, 0, 0]
plt.pie(status_count, labels=status, explode=explode, autopct="%1.1f%%", shadow=True, wedgeprops={'edgecolor': 'black'})
plt.tight_layout()
plt.show()
```

Rides cancelled by Driver: 18%<br>Rides cancelled by Customer: 7%<br>Times when no riders were found: 7%<br>Incomplete Rides: 6%

```python
completed=bk_status["Completed"]
incomplete=bk_status.sum()-completed
sucess_rate=[completed, incomplete]
label=["Complete Rides", "Incomplete Rides"]
plt.pie(sucess_rate, labels=label, autopct="%1.1f%%", shadow=True, wedgeprops={'edgecolor': 'black'})
plt.tight_layout()
plt.show()
```

Total Percentage of completed rides= 62%<br>Total Percentage of incomplete rides= 38%

```python
vec_status=data.groupby(["Vehicle Type", "Booking Status"]).size().unstack()
vec_status["Completion Rate"]=(vec_status["Completed"]/vec_status.sum(axis=1))*100
print(vec_status[["Completed", "Completion Rate"]])
```

The Completion Rate of All vehichles is almost same.

VTAT:Vehicle Turnaround Time i.e. the time it takes for a vehicle to accept/arrive

CTAT: Customer Turnaround Time i.e. the time from booking to trip completion

```python
avg_v_c= data.groupby("Vehicle Type")[["Avg VTAT", "Avg CTAT"]].mean()
print(avg_v_c)
```

```python
total_custom=data["Customer ID"].value_counts()

repeat_custom = (total_custom > 1).sum()
single_custom = (total_custom == 1).sum()

print("Repeat customers:", repeat_custom)
print("One-time customers:", single_custom)
```

Almost all customers have exactly 1 booking.
There's hardly anyone who repeats.

```python
data["Time"]=pd.to_datetime(data["Time"], format="%H:%M:%S")
```

```python
data["Hours"]=data["Time"].dt.hour
x=data["Hours"].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.bar(x.index, x.values, color="#D2042D")
plt.title("Bookings by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Bookings")
plt.xticks(range(24))
plt.show()
```

```python
data["Date"]=pd.to_datetime(data["Date"], format="%Y-%m-%d")
day= data["Date"].dt.day_name().value_counts()
plt.figure(figsize=(10,5))
ordered_days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
day = day.reindex(ordered_days)
plt.bar(day.index, day.values, color="#8fce00")
plt.title("Bookings by Day")
plt.xlabel("Days")
plt.ylabel("Number of Bookings")
plt.show()
```
